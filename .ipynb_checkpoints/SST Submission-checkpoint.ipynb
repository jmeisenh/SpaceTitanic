# Data Wrangling

```python
# split Cabin into deck, room , and side
train = pd.concat([train.Cabin.str.split("/", expand = True).rename(columns = {0:'Deck', 1:'Room', 2:'Side'}),
        train.drop(['Cabin'], axis = 1)], axis = 1
)
# split passenger ID into group and ID
train = pd.concat([train.PassengerId.str.split("_", expand = True).rename(columns = {0:'Group', 1:'Id'}),
        train.drop(['PassengerId'], axis = 1)], axis = 1
)
# Change data types of Group, ID, Room, and Age to int
train[['Group', 'Id', 'Room', 'Age']] = train[['Group', 'Id', 'Room', 'Age']].astype('Int64')
# convert side column to boolean starboard
encode_dict = {
    'Side' : {'P' : False, 'S' : True}
}
train.replace(encode_dict, inplace = True)
train.rename(columns = {'Side' : 'Starboard'}, inplace = True)
train[['Starboard']] = train[['Starboard']].astype('boolean')
```

### Use Existing Statistics To Infer More NAs

#### CryoSleep by HomePlanet

```python
# Europans have a 43.9% Chance of being in cryosleep
# Martians have a 36% Chance of not being in cryosleep
# Terrans have a 31% chance of cryosleep
planets = ['Europa', 'Mars', 'Earth']
cryoPer = [43.9, 36.0, 31.0]
for i, j in zip(planets, cryoPer):
    train.loc[(train['HomePlanet'] == i) & (train['CryoSleep'].isna()), ['CryoSleep']] = random.choices([True, False], weights = (j, 100-j))
```

#### VIP Status by HomePlanet

```python
#Europans have a 6.3% Chance of being a VIP
#Martians have a 3.8% chance of being a VIP
#Terrans have a 0% chance of being a VIP
vipPer = [6.3, 3.8, 0]
for i, j in zip(planets, vipPer):
    train.loc[(train['HomePlanet'] == i) & (train['VIP'].isna()), ['VIP']] = random.choices([True, False], weights = (j, 100-j))
```

#### Deck by HomePlanet

```python
# Europans are on decks A,B,C,D,E, and T
# Martians are on F,E, and D decks
# Terrans are on G,F, and E decks
deckLists = [['A', 'B', 'C', 'D', 'E', 'T'], ['F', 'E', 'D'], ['G', 'F', 'E']]
for i, j in zip(planets, deckLists):
    train.loc[(train['HomePlanet'] == i) & (train['Deck'].isna()), ['Deck']] = random.choices(j)
```

#### Room Number by Deck

```python
# how many people can be in each room on each deck
maxCap = train.groupby(['Deck', 'Room'], as_index = False)['Room'].count().groupby(['Deck']).max().reset_index()
maxCap
```

```python
# Create Room number, Deck dictionary
roomDict = {}
for i in train['Deck'].unique():
    roomDict[i] = (train.loc[train['Deck'] == i]['Room'].unique())
```

```python
# assign random room for each row with NA in room number
for i in train.loc[train['Room'].isna()].index:
    deckNum = train.iloc[i]['Deck']
    # Generate random room already in room list in that deck
    roomRand = random.choices(roomDict[deckNum])[0]
    # find number of passengers in that room
    numInRoom = (train.loc[(train['Deck'] == deckNum) & (train['Room'] == roomRand)].groupby(['Room'])['Room'].count().any())
    # lookup max number of people in rooms on that deck
    roomMax = maxCap.loc[maxCap['Deck'] == deckNum]['Room'].unique()[0]
    if numInRoom < roomMax:
        # assign random room to passenger
        train.iloc[i, 3] = roomRand
    else:
        #generate random rooms until random room is less than room max
        while numInRoom >= roomMax:
            roomRand = random.choices(roomDict[deckNum])[0]
            numInRoom = numInRoom = (train.loc[(train['Deck'] == deckNum) & (train['Room'] == roomRand)].groupby(['Room'])['Room'].count().any())
        # break while loop and assign new random room to passenger
        train.iloc[i, 3] = roomRand
        
    
    
```

```python
# Undummy columns
decklist = [col for col in trainDumKNN if 'Deck_' in col]
trainDumKNN['Destination'] = trainDumKNN[['Destination_PSO J318.5-22', 'Destination_TRAPPIST-1e', 'Destination_55 Cancri e']].idxmax(axis=1).str.replace('Destination_', '')
```

```python
trainDumKNN['Deck'] = trainDumKNN[decklist].idxmax(axis=1).str.replace('Deck_', '')
```

```python
planetList = [col for col in trainDumKNN if 'HomePlanet_' in col]
trainDumKNN['HomePlanet'] = trainDumKNN[planetList].idxmax(axis = 1).str.replace('HomePlanet_', '')
```
