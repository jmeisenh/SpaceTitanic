### Dataset Description

In this competition your task is to predict whether a passenger was transported to an alternate dimension during the Spaceship Titanic's collision with the spacetime anomaly. To help you make these predictions, you're given a set of personal records recovered from the ship's damaged computer system.
File and Data Field Descriptions

   #### train.csv - Personal records for about two-thirds (~8700) of the passengers, to be used as training data.
**PassengerId** - A unique Id for each passenger. Each Id takes the form gggg_pp where gggg indicates a group the passenger is travelling with and pp is their number within the group. People in a group are often family members, but not always.  

**HomePlanet** - The planet the passenger departed from, typically their planet of permanent residence. **CryoSleep** - Indicates whether the passenger elected to be put into suspended animation for the duration of the voyage. Passengers in cryosleep are confined to their cabins.  
**Cabin** - The cabin number where the passenger is staying. Takes the form deck/num/side, where side can be either P for Port or S for Starboard.  
**Destination** - The planet the passenger will be debarking to.  
**Age** - The age of the passenger.  
**VIP** - Whether the passenger has paid for special VIP service during the voyage.  
**RoomService**, **FoodCourt**, **ShoppingMall**, **Spa**, **VRDeck** - Amount the passenger has billed at each of the Spaceship Titanic's many luxury amenities.  
**Name** - The first and last names of the passenger.  
**Transported** - Whether the passenger was transported to another dimension. This is the target, the column you are trying to predict.  

#### Initial Thoughts  
+ Simple categorical data - A Tree model is a good place to start  
+ The **Cabin** variable needs to be split into multiple features with port and starboad converted to binary  
+ **PasengerID** may need to be split into group and ID, possibly even an engineered feature for group size  
+ Luxury amentities could be transformed to weather or not a passenger used a facility and total amount spent
+ Name as it exists is useless for categorizing, however breaking it into first and last name and determing if travling in a family unit is more likely to result in transportation is an experiment to run

```python
# looks like we're going to have to use some imputation methods...how about KNN
# we can use KNN to impute cryosleep, ameneties spending, and age
from sklearn.impute import KNNImputer
```

```python
# should have done this awhile ago but should convert datatypes of columns
for i in ['Group', 'ID', 'Age', 'Room']:
    dfCleaned[i] = dfCleaned[i].astype('Int64')
for i in ['CryoSleep', 'VIP']:
    dfCleaned[i] = dfCleaned[i].astype('boolean')
```

```python
#define categorical columns
catVars = dfCleaned[['HomePlanet', 'Destination', 'Deck', 'Side']]
catDum = pd.get_dummies(catVars, drop_first=True)
catDum.head()
```

```python
# create new KNN imputation df
dfKNN = dfCleaned.drop(['HomePlanet', 'Destination', 'Deck', 'Side', 'FirstName', 'LastName', 'Transported', 'TotalSpent'], axis =1)
dfKNN = pd.concat([dfKNN, catDum], axis =1)
dfKNN
```

```python
# scale data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
dfKNN = pd.DataFrame(scaler.fit_transform(dfKNN), columns = dfKNN.columns)
dfKNN
```

```python
imputer = KNNImputer(n_neighbors= 5)
dfKNN = pd.DataFrame(imputer.fit_transform(dfKNN), columns = dfKNN.columns)
dfKNN.head()
```

```python
# unscale data
dfKNN = pd.DataFrame(scaler.inverse_transform(dfKNN), columns = dfKNN.columns)
# convert CryoSleep back to boolean
dfKNN['CryoSleep'] = dfKNN['CryoSleep'].astype(bool)
```

```python
# convert age to int
dfKNN['Age'] = dfKNN['Age'].astype(int)
# convert VIP to bool
dfKNN['VIP'] = dfKNN['VIP'].astype(bool)
dfKNN
```

```python
# copy new imputed values for cryosleep, age, VIP, and ammeneties to cleaned DF
dfCleaned[['CryoSleep', 'Age', 'VIP', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']] = dfKNN[['CryoSleep', 'Age', 'VIP', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']]
```

```python
# lets try a less black box method like SVM
from sklearn import svm
svm_model = svm.SVC()
```

```python
# Take our cleaned dummified clean train data and minmax scale it
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaleCols = ['Group', 'ID', 'Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 'Room']
boolCols = ['CryoSleep', 'VIP', 'Transported']
dfCleanedDumScaled[scaleCols] = scaler.fit_transform(dfCleanedDum[scaleCols])
dfCleanedDumScaled
```

```python
dfCleanedDumScaled = pd.DataFrame(dfCleanedDumScaled)
dfCleanedDumScaled.columns = ['Group', 'ID', 'CryoSleep', 'Age', 'VIP', 'RoomService', 'FoodCourt',
       'ShoppingMall', 'Spa', 'VRDeck', 'Room', 'Transported',
       'HomePlanet_Europa', 'HomePlanet_Mars', 'Destination_TRAPPIST-1e',
       'Deck_B', 'Deck_C', 'Deck_D', 'Deck_E', 'Deck_F', 'Deck_G', 'Deck_H',
       'Deck_T', 'Side_S']
dfCleanedDumScaled.columns
```

```python
# split data
Xsvm = dfCleanedDumScaled.drop(['Transported'], axis = 1)
Ysvm = dfCleanedDumScaled.iloc[:, 11]
xsvm_train, xsvm_test, ysvm_train, ysvm_test = train_test_split(Xsvm, Ysvm)
```

```python
# define param grid
svm_param_grid = {
    'C': [0.1, 1, 10, 100, 1000],
    'gamma' : [1, 0.1, 0.001, 0.0001],
    'kernel' : ['poly', 'rbf', 'sigmoid']
}
grid_search_svm = GridSearchCV(svm_model, svm_param_grid, cv = 5, n_jobs=-1, verbose=2)
grid_search_svm.fit(xsvm_train, ysvm_train)
```
