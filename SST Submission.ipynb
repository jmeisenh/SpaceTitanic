## Visualize Continous Columns vs Target

```python
# ameneties spending is very skewed towards 0
ameneties = train.iloc[:, 7:12]
for i in ameneties.columns:
    fig = px.histogram(train.loc[train[i] > 100], x = i, color = 'Transported', range_x = [100, 15000],
                      category_orders={
                          'Transported' : [False, True]
                      })
    fig.show()
```

```python
#low spending columns displayed in log y scale for ease of viewing 
for i in ameneties.columns:
    fig = px.histogram(train.loc[train[i] <= 100], x = i, color = 'Transported', log_y=True, nbins = 6, text_auto=True,
                      category_orders={
                          'Transported' : [False, True]
                      })
    fig.update_yaxes(showticklabels = False)
    fig.show()
```

## Conclusions
+ Low spending more correlated with possibility of being transported, especially spending 0
+ Larger spending in shopping mall shows less of a protection against being transported than any other type
